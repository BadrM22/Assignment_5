**Write a function that computes the volume of a sphere given its radius.**
<p>The volume of a sphere is given as $$\frac{4}{3} Ï€r^3$$</p>

```python
import math
def vol(rad):
    return (4/3 * math.pi * rad**3)
```

```python
# Check
vol(2)
```

**Write a function that checks whether a number is in a given range (inclusive of high and low)**

```python
def ran_check(num,low,high):
    if low<=num<=high:
        return f"{num} is in range between {low} and {high}"
    else:
        return f"{num} is not in range between {low} and {high}"
```

```python
# Check
ran_check(5,2,7)
```

If you only wanted to return a boolean:

```python
def ran_bool(num,low,high):
    return True if low<=num<=high else False
```

```python
ran_bool(3,1,10)
```

**Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters.**

    Sample String : 'Hello Mr. Rogers, how are you this fine Tuesday?'
    Expected Output : 
    No. of Upper case characters : 4
    No. of Lower case Characters : 33

HINT: Two string methods that might prove useful: **.isupper()** and **.islower()**

If you feel ambitious, explore the Collections module to solve this problem!

```python
from collections import UserString
def up_low(s):
    upper=0
    lower=0
    for letter in s:
        if letter.isupper():
            upper+=1
        elif letter.islower():
            lower+=1
    string_all=f"Original String : {s}\nNo. of upper case characters : {upper}\nNo. of lower case characters : {lower}" 
    print(string_all)
    
```

```python
s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up_low(s)
```

**Write a Python function that takes a list and returns a new list with unique elements of the first list.**

    Sample List : [1,1,1,1,2,2,3,3,3,3,4,5]
    Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):
    new_lst=[n for n in set(lst)]
    return new_lst
```

```python
unique_list([1,1,1,1,2,2,3,3,3,3,4,5])
```

**Write a Python function to multiply all the numbers in a list.**

    Sample List : [1, 2, 3, -4]
    Expected Output : -24

```python

def multiply(numbers):  
    multi=1
    for num in numbers:
        multi*=num
    return multi


```

```python
multiply([1,2,3,-4])
```

**Write a Python function that checks whether a passed in string is palindrome or not.**

Note: A palindrome is word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

```python
def palindrome(s):
    return True if s == s[::-1] else False
```

```python
palindrome('helleh')
```
